```python
 import numpy as np
 import matplotlib.pyplot as plt
 from tqdm import tqdm
```

```python
def sample_data(dimensions, length, n_points):
  vectors = np.zeros((n_points, dimensions))
  for k in range(n_points):
    for i in range(dimensions):
      vectors[k, i] = np.random.uniform(low=0.0, high=length)

  return vectors

def comp_avr_distance(vectors):
  dists = []
  for k in range(vectors.shape[0]):
    for j in range(vectors.shape[0]):
      dists.append(np.linalg.norm(vectors[k, :] - vectors[j, :]))

  avr_dist = sum(dists) / (len(dists) - vectors.shape[0])
  return avr_dist

def experiment(dims, length, n_points):
  avr_dists = []

  with tqdm(total= max(dims.shape)) as pbar:
    for i in range(max(dims.shape)):
      avr_dists.append(comp_avr_distance(sample_data(
          dimensions=dims[i],
          length=length,
          n_points = n_points
      )))
      pbar.update(1)

  return avr_dists
```

```python
dims = np.arange(1, 800, 1)
ls = experiment(dims = dims, length=1, n_points=100)
plt.plot(dims, ls)
plt.grid()
```
